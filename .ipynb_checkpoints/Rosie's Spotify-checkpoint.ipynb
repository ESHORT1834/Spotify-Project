```python
import os
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy.util as util
import pandas as pd
from matplotlib import pyplot as plt
from pprint import pprint


cid = "e95094f95f8b4a5089263af1484fec78"
secret = "471a83d85bf843c99f368383a7a5c0e4"
username = "128495218"
redirect_uri = "https://example.com/callback/"

os.environ["SPOTIPY_CLIENT_ID"] = cid
os.environ["SPOTIPY_CLIENT_SECRET"] = secret
os.environ["SPOTIPY_REDIRECT_URI"] = redirect_uri
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
scope = 'user-library-read playlist-read-private'
token = util.prompt_for_user_token(username, scope, redirect_uri=redirect_uri)

if token:
    sp = spotipy.Spotify(auth=token)
else:
    print("Can't get token for", username)

print(sp)
```

```python
playlists = sp.user_playlists('spotify')
while playlists:
    for i, playlist in enumerate(playlists['items']):
        print("%4d %s %s" % (i + 1 + playlists['offset'], playlist['uri'],  playlist['name']))
    if playlists['next']:
        playlists = sp.next(playlists)
    else:
        playlists = None
```

```python
rosie_playlist = sp.user_playlist("128495218", "50Hs50rg33Jx4qyr5vUYlY")
tracks = rosie_playlist["tracks"]
songs = tracks["items"]
while tracks['next']:
    tracks = sp.next(tracks)
    for item in tracks["items"]:
        songs.append(item)
ids = [] 
for i in range(len(songs)):
    ids.append(songs[i]['track']['id'])
pprint(ids)
    
    
# print(len(songs))
# print(songs[0]['track']['id'])
# print(rosie_playlist)
pprint(ids)
```

```python
features = []
for i in range(0,len(ids),50):
    audio_features = sp.audio_features(ids)
    for track in audio_features:
        features.append(track)
        features[-1]['target'] = 1
        
pprint(features)
```

```python
r_audio_data = pd.DataFrame(features)
columns = ['analysis_url', 'key', 'mode', 'type', 'uri', 'time_signature', 'duration_ms', 'track_href']
r_audio_data.drop(columns, inplace=True, axis=1)
r_audio_data.head(10)
```

```python
tempo = r_audio_data[r_audio_data['target'] == 1]['tempo']
dance = r_audio_data[r_audio_data['target'] == 1]['danceability']
loudness = r_audio_data[r_audio_data['target'] == 1]['loudness']
speechiness = r_audio_data[r_audio_data['target'] == 1]['speechiness']
valence = r_audio_data[r_audio_data['target'] == 1]['valence']
energy = r_audio_data[r_audio_data['target'] == 1]['energy']
acousticness = r_audio_data[r_audio_data['target'] == 1]['acousticness']
instrumentalness = r_audio_data[r_audio_data['target'] == 1]['instrumentalness']
liveness = r_audio_data[r_audio_data['target'] == 1]['liveness']
```

```python
fig = plt.figure(figsize=(12,8))
plt.title("Rosie's Ideal Song Components")
dance.hist(alpha=0.5, bins=30, label='dance', color ="blue")
valence.hist(alpha=0.5, bins=30, label='valence',color ="green")
energy.hist(alpha=0.5, bins=30, label='energy',color ="purple")
plt.legend(loc='upper right')
```
