```python
kayla_playlist = sp.user_playlist("YOUR_EMAIL", "2hHZqUP3qVt7OLAU1kiDE4")
tracks = kayla_playlist["tracks"]
songs = tracks["items"]
while tracks['next']:
    tracks = sp.next(tracks)
    for item in tracks["items"]:
        songs.append(item)
ids = [] 
for i in range(len(songs)):
    ids.append(songs[i]['track']['id'])
print(ids)
    
    
# print(len(songs))
# print(songs[0]['track']['id'])
# print(kayla_playlist)
print(ids)
```

```python

k_features = []
for i in range(0,len(ids),50):
    audio_features = sp.audio_features(ids)
    for track in audio_features:
        k_features.append(track)
        k_features[-1]['target'] = 1
        
pprint(k_features)
```

```python
k_audio_data = pd.DataFrame(k_features)
k_audio_data
```

```python
columns = ['analysis_url', 'key', 'mode', 'type', 'uri', 'time_signature', 'duration_ms', 'track_href']
k_audio_data.drop(columns, inplace=True, axis=1)
k_audio_data.head(10)
```

```python
plt.scatter(k_audio_data['energy'], k_audio_data['valence'], c='magenta', alpha=0.5)

plt.title("Kayla's Playlist Energy vs Valence")
plt.ylabel("Valence")
plt.xlabel("Energy")
```

```python
k_tempo = k_audio_data[k_audio_data['target'] == 1]['tempo']
k_dance = k_audio_data[k_audio_data['target'] == 1]['danceability']
k_speechiness = k_audio_data[k_audio_data['target'] == 1]['speechiness']
k_valence = k_audio_data[k_audio_data['target'] == 1]['valence']
k_energy = k_audio_data[k_audio_data['target'] == 1]['energy']
k_acousticness = k_audio_data[k_audio_data['target'] == 1]['acousticness']
k_instrumentalness = k_audio_data[k_audio_data['target'] == 1]['instrumentalness']
```

```python
fig = plt.figure(figsize=(12,8))
plt.title("Kayla's Playlist Graph")
k_dance.hist(alpha=.7, bins=30, label='danceability', color ="blue")
k_valence.hist(alpha=.7, bins=30, label='valence',color ="magenta")
k_energy.hist(alpha=.7, bins=30, label='energy',color ="yellow")
plt.legend(loc='upper right')
```
